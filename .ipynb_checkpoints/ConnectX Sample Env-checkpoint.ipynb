# Create ConnectX Environment

```python
from kaggle_environments import evaluate, make, utils
import gym
from DqnConnectX import CNN, ReplayMemory, Dqn

env = make("connectx", debug=True)
```

```python
reward = 0
```

# Create an Agent

```python
brain = Dqn((env.configuration.rows, env.configuration.columns), env.configuration.columns, gamma=0.9)
import numpy as np
reward_windows = []

# This agent random chooses a non-empty column.
def my_agent(observation, configuration):
    import random
    from random import choice
    action = brain.update(reward, observation.board)
    reward_windows.append(np.mean(brain.reward_window))
    selection = action.item()
    return selection if observation.board[selection] == 0 else choice([c for c in range(configuration.columns) if observation.board[c] == 0])
```

# Test the Agent

```python
env.reset()
# Play as the first agent against default "random" agent.
env.run([my_agent, "negamax"])
env.render(mode="ipython", width=500, height=450)
```

```python
# Play as first position against random agent.
import numpy as np

for i in range(100):
    trainer = env.train([None, "negamax"])

    observation = trainer.reset()

    while not env.done:
        my_action = my_agent(observation, env.configuration)
        #print("My Action", my_action)
        observation, reward, done, info = trainer.step(my_action)
        
    if i % 20 == 0:
        print(i)
    
    #print(reward)
    # env.render(mode="ipython", width=100, height=90, header=False, controls=False)
# env.render()
# print(np.array(observation.board).reshape(env.configuration.rows, env.configuration.columns))
# print(done)
```

```python
def mean_reward(rewards):
    return sum(r[0] for r in rewards) / (sum(r[0] + r[1] for r in rewards) + 1)

# Run multiple episodes to estimate its performance.
print("My Agent vs Random Agent:", mean_reward(evaluate("connectx", [my_agent, "random"], num_episodes=10)))
print("My Agent vs Negamax Agent:", mean_reward(evaluate("connectx", [my_agent, "negamax"], num_episodes=10)))
```

```python
import inspect
import os

def write_agent_to_file(function, file):
    with open(file, "a" if os.path.exists(file) else "w") as f:
        f.write(inspect.getsource(function))
        print(function, "written to", file)

write_agent_to_file(my_agent, "submission.py")
```
